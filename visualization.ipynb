```python
import pickle
import os

recording_folder = './recording_data'

# read a pickle file by name from records
def read_pickle(index: int):
    with open(os.path.join(recording_folder, f"record_{index}.pickle"), 'rb') as f:
        return pickle.load(f)
```

```python
import matplotlib.pyplot as plt
from scipy.spatial.transform import Rotation as R

# read a pickle file by name from records
data = read_pickle(2)

# plot the data of bat in data
bat_pos = []
bat_rot = []

for i, step_data in enumerate(data):
    bat_pos.append(step_data['obs']['bat1']['xpos'])
    quat = step_data['obs']['bat1']['xquat']
    quat = [quat[1], quat[2], quat[3], quat[0]]
    rot = R.from_quat(quat).as_euler('xyz', degrees=True)
    bat_rot.append(R.from_quat(quat).as_euler('xyz', degrees=True))

# plot the data of bat_pos in each axis
plt.plot(bat_pos)
plt.show()

```

```python
plt.plot(bat_rot)
plt.show()
```
